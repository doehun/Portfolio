## Stopwords

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words.extend(['The', 'Red', 'Hat', 'To', 'With', 'with', 'using', 'Using', 'This', 'use' , 'It', 'it', 'More', 'Explore',
                 'Resource', 'Resources', 'Details', 'To', 'information', 'need', 'needs', 'world', 'leading', 'provider', 'solutions', 'See'])
stop_words = set(stop_words)
```

## Simple Counts - Uni, Bi, and Trigrams

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(input = 'content', min_df = 5, stop_words = stop_words, ngram_range = (2,2), lowercase = False, analyzer = 'word' ) 
dtm = vectorizer.fit_transform(data['Descriptions'])
```

```python
#vectorizer.get_feature_names()
```

## LDA And LDAvis

```python
from sklearn import decomposition

# 'Optimal' LDA hyperparameters - I provided a history of hyperparameters below
num_topics = 13
doc_topic_prior = .4
topic_word_prior = .01
max_iter = 200
evaluate_every = 10

clf = decomposition.LatentDirichletAllocation(n_topics = num_topics, doc_topic_prior = doc_topic_prior, topic_word_prior = topic_word_prior, max_iter = max_iter, evaluate_every = evaluate_every)
doctopic = clf.fit_transform(dtm) 
```

```python
# A lower perplexity score indicates better generalization performance
clf.perplexity(dtm)
```

```python
clf.score(dtm)
```

```python
import pyLDAvis.sklearn
import numpy

# You can use other metrics of multidimensional scaling. mds = 'PCoA' (default), mds = 'tsne', mds = 'mmds'
# This doesn't change the LDA, just the visualization of Intertopic Distance Map

ngram_vis = pyLDAvis.sklearn.prepare(clf, dtm, vectorizer, mds='PCoA') 
```
